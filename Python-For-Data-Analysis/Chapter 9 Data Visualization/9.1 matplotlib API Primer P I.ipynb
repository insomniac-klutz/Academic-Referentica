Creating  a  figure  with  a  grid  of  subplots  is  a  very  common  task,  so  matplotlibincludes a convenience method, plt.subplots, that creates a new figure and returnsa NumPy array containing the created subplot objects:

```python
fig, axes = plt.subplots(2, 3)
```

This  is  very  useful,  as  the  axes  array  can  be  easily  indexed  like  a  two-dimensionalarray;  for  example,  axes[0, 1].  You  can  also  indicate  that  subplots  should  have  thesame  x-  or  y-axis  using  sharex  and  sharey,  respectively.  This  is  especially  usefulwhen youâ€™re comparing data on the same scale

<b>Adjusting the spacing around subplots

You  canchange the spacing using the subplots_adjust method on Figure objects, also availâ€able as a top-level function:

    subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

wspace and hspace controls the percent of the figure width and figure height, respecâ€tively, to use as spacing between subplots

```python
ig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        
        axes[i, j].hist(np.random.randn(500), bins=50, color='b', alpha=0.5)
        plt.subplots_adjust(wspace=i*i, hspace=j*j)
```

<h3>Colors, Markers, and Line Styles

```python
x=np.random.randn(100)
y=np.random.randn(100)
figf=plt.figure()
o=plt.plot(x, y,linestyle='--',color= 'b')
o
```

Line  plots  can  additionally  have  markers  to  highlight  the  actual  data  points.  Sincematplotlib  creates  a  continuous  line  plot,  interpolating  between  points,  it  can  occaâ€sionally  be  unclear  where  the  points  lie.  The  marker  can  be  part  of  the  style  string,which must have color followed by marker type and line style 

```python
figf=plt.figure()
plt.plot(x, y,linestyle='--',color= 'b',marker='o')
o
```

For  line  plots,  you  will  notice  that  subsequent  points  are  linearly  interpolated  bydefault. This can be altered with the drawstyle option

```python
figf=plt.figure()
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```
