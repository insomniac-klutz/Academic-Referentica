<b> Random Walks using random module</b>

```python
%%time
import matplotlib.pyplot as plt
import random as rdm
import numpy as np

position=0
map=[position]
steps=2121

for i in range(steps):
    step=1 if rdm.randint(0,1) else -1
    position+=step
    map.append(position)

plt.plot(map[:])
plt.title("Image plot of Random Walk values")
```

<b> Random Walks using numpy module</b>

```python
%%time
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
#print(draws)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
plt.plot(walk)
```

```python
(np.abs(walk) >= 10).argmax()
```

```python
(np.abs(walk) >= 10).sum()
```

<b> Multiple Random Walks using numpy module</b>

```python
%%time
nwalks=5
nsteps = 2102
draws = np.random.randint(0, 2, size=(nwalks,nsteps))
#print(draws)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum(1)
#print(walk)
for i in enumerate(walk):
    plt.plot(walk[i])
    if(i==4):
        break
```
